```python
import tkinter as tk
import random

# Function to determine the winner
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (
        (user_choice == "rock" and computer_choice == "scissors")
        or (user_choice == "scissors" and computer_choice == "paper")
        or (user_choice == "paper" and computer_choice == "rock")
    ):
        return "You win!"
    else:
        return "Computer wins!"

# Function to handle user's choice
def play(user_choice):
    computer_choices = ["rock", "paper", "scissors"]
    computer_choice = random.choice(computer_choices)

    result = determine_winner(user_choice, computer_choice)

    result_label.config(text=f"Computer chose {computer_choice}. {result}", fg="blue")

    # Update scores (optional)
    if result == "You win!":
        user_score.set(user_score.get() + 1)
        user_score_label.config(text=f"Your Score: {user_score.get()}", fg="green")
    elif result == "Computer wins!":
        computer_score.set(computer_score.get() + 1)
        computer_score_label.config(text=f"Computer Score: {computer_score.get()}", fg="red")

# Create the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors Game")
root.geometry("400x200")  # Set window size

# Label and buttons
instruction_label = tk.Label(root, text="Choose rock, paper, or scissors:", fg="purple", font=("Helvetica", 14))
instruction_label.pack(pady=10)

button_frame = tk.Frame(root)
button_frame.pack()

rock_button = tk.Button(button_frame, text="Rock", width=10, height=2, command=lambda: play("rock"), bg="gray")
paper_button = tk.Button(button_frame, text="Paper", width=10, height=2, command=lambda: play("paper"), bg="lightblue")
scissors_button = tk.Button(button_frame, text="Scissors", width=10, height=2, command=lambda: play("scissors"), bg="lightgreen")

rock_button.grid(row=0, column=0, padx=10)
paper_button.grid(row=0, column=1, padx=10)
scissors_button.grid(row=0, column=2, padx=10)

result_label = tk.Label(root, text="", fg="blue", font=("Helvetica", 12))
result_label.pack(pady=10)

# Optional: Score tracking
user_score = tk.IntVar()
computer_score = tk.IntVar()
user_score.set(0)
computer_score.set(0)

user_score_label = tk.Label(root, text="Your Score: 0", fg="green", font=("Helvetica", 12))
computer_score_label = tk.Label(root, text="Computer Score: 0", fg="red", font=("Helvetica", 12))
user_score_label.pack()
computer_score_label.pack()

# Run the application
root.mainloop()
```
